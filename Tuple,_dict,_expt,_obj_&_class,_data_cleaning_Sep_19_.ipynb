```python
# Declear a Dictionary
dict= {"name":"ken", "age": [52.2, "72"], "hoobies": ["singing", "running"]}
print(dict)
```

```python
#Using for loop to display only values in a dict
for values in dict.values():
  print(values)
```

```python
#How to get the value using a dict key
x= dict.get("age")
print(x)
```

```python
#Another method of acessing value using key
x= dict["age"]
print(x)
```

```python
# Declaration of a class
#from os import name
class person:
    def __init__(self, name, age):
      self.name= name
      self.age= age
# Declaration of an object
p1= person("Jhon", 32)
p2= person("peter", 30)
#Display the values of obejects along with a variable
print("name of a person:", p1.name)
print("age of a person:", p1.age)
#Calling person 2
print("name of a person:", p2.name)
print("age of a person:", p2.age)
```
