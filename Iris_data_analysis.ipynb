```python
@interact # Interact helps to interact with the dataframe and manupulate each column
def show_set(colunm= 'SepalLengthCm', x=5):
  return Iris.loc[Iris[colunm] > x] # defing a custum function
```

```python
plt.figure(figsize = (6, 4))
plt.hist(Iris['SepalLengthCm'],
        bins = 20, color = "y")
plt.xlabel("SepalLengthCm")
plt.ylabel("count")
plt.title("SPLC Histogram") #SPLC(SepalLengthCm)
```

```python
plt.figure(figsize = (6, 4))
plt.hist(Iris['SepalWidthCm'],
        bins = 20, color = "b")
plt.xlabel("SepalWidthCm")
plt.ylabel("count")
plt.title("SPWC Histogram") #SPLC(SepalWidthCm)
```

```python
#This code creates a subplot(2 fig beside each other)
fig, ax = plt.subplots(1,2,
                       figsize = (14,6)) #(This line codes the dimension of the plot, "1,2"(1 row, 2 Column i.e plot is beside eachother)
               #                            "2,1" (2rows, 1 column i.e plot on top of eachother) figsize "15,6"(15 wide, 6 high))
Iris.plot(x = 'SepalLengthCm', y ='SepalWidthCm',
          kind = "scatter", ax = ax[0], #NB(ax: is the axis, ax[0]: specifies the position septal plot will be (left or right) like an array[0,1], 0 will come before 1)
          sharex =False, sharey = False, #NB (What diffrence does this line of code makes on the plot?,I tried coding T inplace F the fig came back thesame  )
          label= 'Septal', color= 'r')
Iris.plot(x = 'PetalLengthCm', y ='PetalWidthCm',
          kind = "scatter", ax = ax[1],
          sharex =False, sharey = False,
          label= 'Petal', color= 'b')
plt.suptitle("petal vs septal")
plt.show()

#plt.scatter(Iris['SepalLengthCm'],Iris['SepalWidthCm'])
#plt.xlabel('SepalLengthCm')
#plt.ylabel('SepalWidthCm')
#plt.title('Sepal')
#plt.show()

#plt.figure(figsize=(6,4))
#plt.scatter(Iris['PetalLengthCm'],Iris['PetalWidthCm'])
#plt.xlabel('PetalLengthCm')
#plt.ylabel('PetalWidthCm')
#plt.title('Petal')
#plt.show()
```

```python
# Using the interact package to modify plot
def plot(Species):
  data = Iris[Iris.Species==Species] #Defining function plot(species).Code says: "plot(species) is defined as data, which is the specie variable in the Iris dataframe"

  data.plot.scatter('SepalLengthCm', # code says: plot a scatter plot of 'septallenght vs spetalwidt' of rows in Iris dataframe that has 'Irisvirginica' specie alone
                    'SepalWidthCm',
  label= 'Sepal/species', color= 'b')
interact(plot, kind= 'scatter',
         Species= 'Iris-virginica')
plt.show()
```
